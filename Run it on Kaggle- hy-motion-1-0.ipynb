```python
import os

# 1. Force Hugging Face to use the massive /tmp partition for caching downloads
os.environ["HF_HOME"] = "/tmp/hf_cache"

# 2. Create the checkpoint directory inside /tmp
!mkdir -p /tmp/ckpts/tencent
!mkdir -p /tmp/ckpts/clip-vit-large-patch14
!mkdir -p /tmp/ckpts/Qwen3-8B

# 3. Symlink it to your current folder so the project code can find it seamlessly
!ln -s /tmp/ckpts ckpts
# 1. Download the Lite motion generation weights
!./myenv/bin/hf download tencent/Hy-Motion-1.0 --include "HY-Motion-1.0-Lite/*" --local-dir ckpts/tencent

# 2. Download the CLIP Text Encoder
!./myenv/bin/hf download openai/clip-vit-large-patch14 --local-dir ckpts/clip-vit-large-patch14/

# 3. Download the Qwen Text Encoder Base
!./myenv/bin/hf download Qwen/Qwen3-8B --local-dir ckpts/Qwen3-8B
print("Storage successfully rerouted to Kaggle's high-capacity drive!")
```

```python
launcher_code = """
import os
import sys
import transformers
import torch

# Prevent massive Prompter LLM downloads
os.environ["DISABLE_PROMPT_ENGINEERING"] = "True"
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"
os.environ["USE_HF_MODELS"] = "0"

orig_from_pretrained = transformers.AutoModel.from_pretrained
def patched_from_pretrained(model_id, *args, **kwargs):
    if "Qwen" in str(model_id) or "8B" in str(model_id):
        local_path = "/tmp/ckpts/Qwen3-8B"
        print(f">>> [VRAM Shield] Loading Qwen in 4-bit on P100...")
        kwargs["load_in_4bit"] = True
        kwargs["device_map"] = {"": 0} 
        kwargs["torch_dtype"] = torch.float16
        model = orig_from_pretrained(local_path, *args, **kwargs)
        return model
    return orig_from_pretrained(model_id, *args, **kwargs)

orig_causal = transformers.AutoModelForCausalLM.from_pretrained
def patched_causal(model_id, *args, **kwargs):
    if "Qwen" in str(model_id) or "8B" in str(model_id):
        local_path = "/tmp/ckpts/Qwen3-8B"
        print(f">>> [VRAM Shield] Loading CausalLM in 4-bit on P100...")
        kwargs["load_in_4bit"] = True
        kwargs["device_map"] = {"": 0}
        kwargs["torch_dtype"] = torch.float16
        model = orig_causal(local_path, *args, **kwargs)
        return model
    return orig_causal(model_id, *args, **kwargs)

transformers.AutoModel.from_pretrained = patched_from_pretrained
transformers.AutoModelForCausalLM.from_pretrained = patched_causal

if len(sys.argv) > 1 and sys.argv[1] == "cli":
    sys.argv.pop(1)
    import local_infer
    local_infer.main()
else:
    with open("gradio_app.py", "r") as f:
        exec(f.read(), globals())
"""
with open("run_hymotion.py", "w") as f:
    f.write(launcher_code.strip())

# Make sure we actually have the inference script
!wget -q -O local_infer.py https://raw.githubusercontent.com/Tencent-Hunyuan/HY-Motion-1.0/master/local_infer.py

print("P100 Launcher built and ready!")
```
